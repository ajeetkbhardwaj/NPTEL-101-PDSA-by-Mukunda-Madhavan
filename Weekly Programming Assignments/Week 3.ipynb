```python
"""  
Here are some examples of how your function should work.

 
  >>> leftrotate([[1,2],[3,4]])
  [[2, 4], [1, 3]]

  >>> leftrotate([[1,2,3],[4,5,6],[7,8,9]])
  [[3, 6, 9], [2, 5, 8], [1, 4, 7]]

  >>> leftrotate([[1,1,1],[2,2,2],[3,3,3]])
  [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
"""
```

```python
# Contracting function
def contracting(l):
  for i in range(0,len(l)-3):
    if (abs(l[i+2]-l[i+1]) < abs(l[i+1]-l[i])):
      continue
    else:
      return(False)
  return(True)


# Counting function
def counthv(l):
  hc=0
  vc=0
  if len(l) > 2:
    for i in range(1,len(l)-1):
      if l[i] > l[i-1] and l[i] > l[i+1]:
        hc+=1
      elif l[i] < l[i-1] and l[i] < l[i+1]:
        vc+=1
  return [hc,vc]


# Rotation function
def leftrotate(m):
    b=[]
    for i in range(len(m)):
        b.append([])
        for j in range(len(m)):
            b[i].append(m[j][len(m)-i-1])
    return(b)
      
     
import ast

def parse(inp):
  inp = ast.literal_eval(inp)
  return (inp)

fncall = input()
lparen = fncall.find("(")
rparen = fncall.rfind(")")
fname = fncall[:lparen]
farg = fncall[lparen+1:rparen]

if fname == "contracting":
  arg = parse(farg)
  print(contracting(arg))

if fname == "counthv":
  arg = parse(farg)
  print(counthv(arg))

if fname == "leftrotate":
  arg = parse(farg)
  savearg = arg
  ans = leftrotate(arg)
  if savearg == arg:
    print(ans)
  else:
    print("Side effect")
```
